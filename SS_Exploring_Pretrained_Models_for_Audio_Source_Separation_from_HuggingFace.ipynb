**Description**: A Proof of Concept (PoC) project to evaluate the performance of pre-trained speech separation models from Hugging Face. The goal is to successfully separate overlapped speech into distinct audio tracks for 2 speakers.

**Tools Used**: Python, Speechbrain, Asteroid.

**Key Learning**: Understanding how to load and deploy models from the HuggingFace.

# speechbrain

```python
!pip install speechbrain
```

```python
from speechbrain.pretrained import SepformerSeparation as separator
import torchaudio

model_speech = separator.from_hparams(source="speechbrain/sepformer-libri2mix")

speech = model_speech.separate_file(path='speechbrain/sepformer-wsj02mix/test_mixture.wav')

torchaudio.save("speechbrain_1.wav", speech[:, :, 0], 8000)
torchaudio.save("speechbrain_2.wav", speech[:, :, 1], 8000)
```

# asteroid

```python
!pip install asteroid
```

```python
data, sr = torchaudio.load('/content/audio_cache/test_mixture.wav')
```

```python
from asteroid.models import BaseModel

model_JorisCos = BaseModel.from_pretrained("JorisCos/ConvTasNet_Libri2Mix_sepnoisy_8k")

JorisCos = model_JorisCos.separate(data)

torchaudio.save("JorisCos_1.wav", JorisCos[:, 0, :], 8000)
torchaudio.save("JorisCos_2.wav", JorisCos[:, 1, :], 8000)
```
