**Description**: A two-stage classification project using the Random Forest algorithm on the Iris dataset. The project compares model performance between using the full feature set and a reduced set of the Top-2 most influential features identified through feature importance analysis.

**Tools Used**: Python, Pandas, Numpy, Scikit-Learn, Tensorflow.

**Key Learning**: Using 'model.feature_importances_' to identify and select key features that significantly impact the model's predictive power. Plotting feature importance scores and decision surfaces for the selected features.

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
import seaborn as sns; sns.set_theme()
```

```python
#Loading the iris dataset
iris = load_iris()
X, y = iris.data, iris.target
X,y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#RandomForest model
model = RandomForestClassifier(n_estimators=30)
#model = DecisionTreeClassifier(criterion = 'entropy', max_depth=3,random_state= 42)
```

```python
#Train the model
model.fit(X_train, y_train)
```

```python
global_exp = pd.Series(model.feature_importances_)
global_exp.sort_values(ascending=False).plot.bar()
plt.ylabel("Feature Importance")
plt.title("Feature Importance - Global Explanations")
plt.show()
```

```python
#Predictions on the test data
y_pred = model.predict(X_test)
```

```python
#Evaluating the model's performance
accuracy = accuracy_score(y_test, y_pred)
print("Decision Tree: {:.2f}%".format(accuracy * 100))
```

```python
#Making the Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
print("Accuracy :",accuracy_score(y_test, y_pred))
print()
ax = sns.heatmap(confusion_matrix(y_test, y_pred),annot=True, fmt="d")
```
