```python
import numpy as np
import pandas as pd
```

```python
iris = pd.read_csv('Iris.csv')
```

```python
iris
```

## **Preprocessing**

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

X = iris.drop('Species', axis=1)
X = X.drop('Id', axis=1)
y = iris['Species']

le = LabelEncoder()
y = le.fit_transform(iris['Species'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

## **Train and test model**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

#Euclidean Distance  euclidean
#Manhattan Distance  cityblock
#Minkowski Distance (with p=3)  minkowski
#Cosine Similarity  cosine
#Chebyshev Distance  chebyshev

A = ['euclidean','cityblock','minkowski','cosine','chebyshev']
Y = []
for i in A:
  knnClassifier = KNeighborsClassifier(n_neighbors = 3, metric = i)
  knnClassifier.fit(X_train, y_train)
  Y.append(knnClassifier.predict(X_test))

Y = np.array(Y).T
y_pred_euclidean , y_pred_manhattan , y_pred_minkowski , y_pred_cosine , y_pred_chebyshev = Y[:,0] , Y[:,1] , Y[:,2] , Y[:,3] , Y[:,4]

df = pd.DataFrame(
    {
        "Distance Metric":['euclidean','manhattan','minkowski','cosine','chebyshev'],
        "Accuracy":[accuracy_score(y_pred_euclidean,y_test),accuracy_score(y_pred_manhattan,y_test),accuracy_score(y_pred_minkowski,y_test),accuracy_score(y_pred_cosine,y_test),accuracy_score(y_pred_chebyshev,y_test)]
    })
df
```

# **Part 2: K-means Clustering**

```python
X = iris
X = X.drop('Id', axis=1)
X_0 = (X[X['Species']=='Iris-setosa'].drop('Species', axis=1))
X_1 = (X[X['Species']=='Iris-versicolor'].drop('Species', axis=1))
X_2 = (X[X['Species']=='Iris-virginica'].drop('Species', axis=1))

centroid0 = np.array(np.sum(X_0,axis=0)/len(X_0))
centroid1 = np.array(np.sum(X_1,axis=0)/len(X_1))
centroid2 = np.array(np.sum(X_2,axis=0)/len(X_2))

X_0 = X_0.to_numpy()
X_1 = X_1.to_numpy()
X_2 = X_2.to_numpy()


def euclidean_distance(u,v):
  return (np.sum((u-v)**2))**0.5

distortion_euclidean = 0
for x in X_0:
  distortion_euclidean += euclidean_distance(x,centroid0)
for x in X_1:
  distortion_euclidean += euclidean_distance(x,centroid1)
for x in X_2:
  distortion_euclidean += euclidean_distance(x,centroid2)

def manhattan_distance(u,v):
  return np.sum(np.abs(u-v))

distortion_manhattan = 0
for x in X_0:
  distortion_manhattan += manhattan_distance(x,centroid0)
for x in X_1:
  distortion_manhattan += manhattan_distance(x,centroid1)
for x in X_2:
  distortion_manhattan += manhattan_distance(x,centroid2)

def minkowski_distance(u,v):
  return np.sum(np.abs(u-v)**3)**(1/3)

distortion_minkowski = 0
for x in X_0:
  distortion_minkowski += minkowski_distance(x,centroid0)
for x in X_1:
  distortion_minkowski += minkowski_distance(x,centroid1)
for x in X_2:
  distortion_minkowski += minkowski_distance(x,centroid2)

def cosine_distance(u,v):
  return np.dot(u,v)/(np.linalg.norm(u)*np.linalg.norm(v))

distortion_cosine = 0
for x in X_0:
  distortion_cosine += cosine_distance(x,centroid0)
for x in X_1:
  distortion_cosine += cosine_distance(x,centroid1)
for x in X_2:
  distortion_cosine += cosine_distance(x,centroid2)

def chebyshev_distance(u,v):
  return np.max(np.abs(u-v))

distortion_chebyshev = 0
for x in X_0:
  distortion_chebyshev += chebyshev_distance(x,centroid0)
for x in X_1:
  distortion_chebyshev += chebyshev_distance(x,centroid1)
for x in X_2:
  distortion_chebyshev += chebyshev_distance(x,centroid2)

df = pd.DataFrame(
    {
        "Distance Metric":['euclidean','manhattan','minkowski','cosine','chebyshev'],
        "Distortion":[distortion_euclidean,distortion_manhattan,distortion_minkowski,distortion_cosine,distortion_chebyshev]
    })
df
```
