**Description**: A step-by-step implementation of the K-Means Clustering algorithm, developed from scratch using Python.

**Tools Used**: Python, Matplotlib (for data visualization and scatter plots).

**Key Learning**: Understanding the mathematical foundations of K-Means Clustering by translating equations into functional code.

```python
import matplotlib.pyplot as plt

x = [-1, -2, -2, -1,  0,  1,  1,  3,  3,  4,  4]
y = [ 6,  4, -2, -1,  2, -3,  5, -2,  6,  0,  2]

mew_x = [2, 2]
mew_y = [-1, 4]

plt.scatter(x, y, color = 'blue')
plt.scatter(mew_x, mew_y, color = 'red')
plt.grid(True)
plt.axhline(color = 'black')
plt.axvline(color = 'black')
plt.show()
```

```python
def distance(x1,x2):
  p = []
  for i in range(len(x1)):
    d = x1[i] - x2[i]
    p.append(d)
  return p

def add(x1,x2):
  p = []
  for i in range(len(x1)):
    d = x1[i] + x2[i]
    p.append(d)
  return p

def div(x1,x):
  p = []
  for i in range(len(x1)):
    if type(x) == list:
      d = x1[i]/x[i]
    else:
      d = x1[i]/x
    p.append(d)
  return p

def norm(X):
  p = 0
  for i in range(len(X)):
    p += X[i]**2
  return p**0.5

X = [[-1, 6], [-2, 4], [-2, -2], [-1, -1], [0, 2], [1, -3], [1, 5], [3, -2], [3, 6], [4, 0], [4, 2]]
Mew1 = [2, -1]
Mew2 = [2,  4]

print(f'The initial cluster centroids are {Mew1} and {Mew2}')

flag = True
while flag == True:
  C = []
  C1 = []
  C2 = []
  for i in range(len(X)):
    if norm(distance(X[i],Mew1)) < norm(distance(X[i],Mew2)):
      M = 1
      C1.append(M)
    else:
      M = 2
      C2.append(M)
    C.append(M)
  sum_CX1 = [0,0]
  sum_CX2 = [0,0]
  for i in range(len(C)):
    if C[i] == 1:
      sum_CX1 = add(sum_CX1,X[i])
    else:
      sum_CX2 = add(sum_CX2,X[i])

  if (div(div(sum_CX1,len(C1)),Mew1)[0] > 1 or div(div(sum_CX1,len(C1)+1),Mew1)[1] > 1) or div(div(sum_CX2,len(C2)+1),Mew2)[0] > 1 or div(div(sum_CX2,len(C2)+1),Mew2)[1] > 1:
    flag = True
  else:
    flag = False

  Mew1 = div(sum_CX1,len(C1))
  Mew2 = div(sum_CX2,len(C2))
  print(Mew1,Mew2)
```

```python
mew_new_x = Mew1
mew_new_y = Mew2

plt.scatter(x, y, color = 'blue')
plt.scatter(mew_x, mew_y, color = 'black')
plt.scatter(mew_new_x, mew_new_y, color = 'red')
plt.grid(True)
plt.axhline(color = 'black')
plt.axvline(color = 'black')
plt.show()
```
