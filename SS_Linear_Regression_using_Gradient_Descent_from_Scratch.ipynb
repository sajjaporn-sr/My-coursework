**Description**: An implementation of Simple Linear Regression using the Gradient Descent optimization algorithm from scratch. This project demonstrates how to iteratively update model parameters ($\beta_0$ and $\beta_1$) based on the error derivatives to minimize the loss.

**Tools Used**: Python, Numpy, Matplotlib.

**Key Learning**: Implementing parameter update rules based on Gradient Descent equations. Calculating and monitoring Mean Squared Error (MSE) to ensure model convergence.

Define **f(x) = $\beta_0$ + $x\beta_1$**

```python
def f(x, beta0, beta1):
    return beta0 + (beta1 * x)
```

Define **X = [1,2,3,4,5,6,7]**

```python
import numpy as np

X = np.array([1,2,3,4,5,6,7])
```

Compute **f(x)** where x in X

```python
beta0 = 1
beta1 = 1
f(X, beta0, beta1)
```

Compute error of **y and f(x)** where **y = [3,2,4,5,4,6,6]** (from the picture)

```python
y = np.array([3,2,4,5,4,6,6])

error = y - f(X, beta0, beta1)
error
```

Define **learning rate lr = 0.01** and adjust $\beta_0$ and $\beta_1$

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([1,2,3,4,5,6,7])
y = np.array([3,2,4,5,4,6,6])

lr = 0.001
epoch = 400

beta0 = 1
beta1 = 1
MSE = []

for i in range(epoch):
  error = f(X, beta0, beta1) - y
  beta0 = beta0 - lr * error.mean()
  beta1 = beta1 - lr * (error * X).mean()

  mse = (error**2).mean()
  MSE.append(mse)
  if i%50 == 0:
    print(f'epoch = {i}, mse = {mse}')

plt.plot(MSE)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x2 = np.linspace(0,10,100)
plt.scatter(X,y)
plt.plot(x2,f(x2, 1, 1), label='initial line', color='orange')
plt.plot(x2,f(x2, beta0, beta1), label='adjusted line', color='green')
plt.legend()
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```
